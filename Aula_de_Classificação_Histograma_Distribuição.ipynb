# Aula de Classificação
Histograma e Distribuição

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('base_prob.csv')
df
```

###Construindo um histograma com as idades

```python
#Chamo a função histograma (hist) que está dentro da biblioteca matplotlib.pyplot (plt)
plt.hist(df.age);
```

#### Por padrão a função hist irá criar 10 colunas (bins).

```python
#Posso alterar o tamanho que aparece o histograma.(Largura, altura)em polegadas.
plt.figure(figsize=(15,5))
plt.hist(df.age);
```

```python
#Posso determinar o intervalo que desejo de colunas. Utilizando agora a biblioteca numpy e a função arange e delimitando o intervalo de interesse.
plt.figure(figsize= (15,5))
plt.hist(df.age, bins= np.arange(0,101,1));
```

```python
#Para melhorar a visualização do histograma posso utilizar a função rwidth que irá propiciar um intervalo entre as colunas.
plt.figure(figsize= (15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= 0.8);
```

```python
#Para colocar um título nos eixos usa-se a função xlabel e ylabel.
plt.figure(figsize= (15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= 0.8)
plt.xlabel("Idade")
plt.ylabel("Contagem de pessoas");
```

```python
#Aumentando as letras dos títulos dos eixos.
plt.figure(figsize= (15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth=0.8)
plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Contagem de pessoas", fontsize= 16);
```

```python
#Podemos aumentar o tamanho dos número dos eixos.
plt.figure(figsize= (15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth=0.8)
plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Contagem de pessoas", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16);
```

```python
#Adicionando título ao histograma
plt.figure(figsize=(15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8)
plt.xlabel("Idade", fontsize=16)
plt.ylabel("Contagem de pessoas", fontsize= 16)
plt.xticks(fontsize=16)
plt.yticks(fontsize= 16)
plt.title("Histograma das Idades", fontsize = 20);
```

```python
#Alterando a cor das colunas no histograma.
plt.figure(figsize= (15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8, color= "R")
plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Contagem de pessoas", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.title("Histograma de Idade", fontsize= 20);

```

```python
#Para facilitar a visualização dos dados podemos colocar as linhas de fundo.
plt.figure(figsize=( 15, 5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= 0.8, color="r")
plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Contagem de pessoas", fontsize= 16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("Histograma das Idades", fontsize= 20)
plt.grid();
```

##Distribuição de probabilidade

```python
#Dentro da própria função do hitograma tem a distribuição da probabilidade.(Density)
plt.figure(figsize=(15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8, color= "g", density= True)
plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Probabilidade/Idade", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.title("Distribuição da Probabilidade", fontsize= 20)
plt.grid();
```

```python
# Destacando um ponto de interesse dentro do histograma.(axvline) linha vertical paralela ao eixo y.
plt.figure(figsize=(15,5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8, color= "g", density= True)
#Escolhendo a idade 29
plt.axvline(29)

plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Probabiliadade/Idade", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.title("Distribuição da Probabilidade", fontsize= 20)
plt.grid();
```

```python
#Por padrão a linha gerada é da cor azul, mas podemos alterar.
plt.figure(figsize=( 15, 5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8, color="g", density= True)
plt.axvline(29, color="k")

plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Probabilidade/Idade", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.title("Distribuição da Probabilidade", fontsize= 20)
plt.grid();
```

```python
#Adicionando uma linha a partir do eixo y também. (axhline)Linha horizontal paralela ao eixo x.
plt.figure(figsize= ( 15, 5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8, color= "g", density= True)
plt.axvline(29, color= "k")
plt.axhline(0.0125, color="k")

plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Probabilidade/Idade", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.title("Distribuição da Probabilidade", fontsize= 20)
plt.grid();
```

```python
#Colocando mais valores no eixo y.
plt.figure(figsize=( 15, 5))
plt.hist(df.age, bins= np.arange(0, 101, 1), rwidth= .8, color="g", density= True)
plt.axvline(29, color= "k")
plt.axhline(0.0125, color= "k")

plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Probabilidade/Idade", fontsize= 16)
plt.title("Distribuição da Probabilidade", fontsize= 20)
plt.xticks(fontsize= 16)
plt.yticks([0, 0.005, 0.010, 0.0125, 0.015, 0.020, 0.025], fontsize= 16)
plt.grid();
```

###Distribuição de probabilidade conjunta

```python
#Para a probabilidade conjunta utilizaremos o hist2d da biblioteca matplotlib.pyplot.
plt.figure(figsize=(15,5))
plt.hist2d(df.age, df.MonthlyIncome, bins= 30, density= True)

plt.xlabel("Idade", fontsize= 16)
plt.ylabel("Probabilidade/Idade", fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.grid()
```

```python
#Adicionando a legenda de cores no histograma (colorbar)
plt.figure(figsize= (15, 5))
plt.hist2d(df.age, df.MonthlyIncome, bins= 30, density= True)
plt.colorbar()

plt.xlabel("Idade", fontsize=16)
plt.ylabel("Renda Mensal",fontsize= 16)
plt.xticks(fontsize= 16)
plt.yticks(fontsize= 16)
plt.title("Distribuição de Renda por Idade", fontsize= 20);

```

```python
#Adicionando dois gráficos na mesma figura, dimenciono o tamanho da minha figira depois utilizo a função subplot
#e passo a localização que desejo. Ao lançar a função subplot indico qual a quantidade de linhas e colunas na 
#figura que pretendo criar.
#subplot(1,2,2) significa 1 linha com 2 colunas e este elemento ficará na coluna 2.
plt.figure(figsize= (15, 5))

plt.subplot(1,2,1)
plt.hist(df.MonthlyIncome, bins=30, color="y", density= True)
plt.xlabel("Renda Mensal", fontsize= 14)
plt.ylabel("Probabilidade/Renda", fontsize= 14)
plt.title("Distribuição da Probabilidade", fontsize= 18)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)


plt.subplot(1,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins= 30, density= True)
plt.colorbar()
plt.xlabel("Idade", fontsize= 14)
plt.ylabel("Renda Mensal", fontsize= 14)
plt.title("Distribuição de Renda por Idade", fontsize= 18)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14);

```

```python
#Transformando as colunas verticais em barras horizontais.
plt.figure(figsize= (15, 5))

plt.subplot(1,2,1)
plt.hist(df.MonthlyIncome, bins= 30, color= "y", density= True, orientation= "horizontal", rwidth= .8)
plt.xlabel("Renda Mensal", fontsize= 14)
plt.ylabel("Probabilidade/Renda", fontsize= 14)
#Inclinando os números do eixo x para melhorar a visualização
plt.xticks(fontsize= 14, rotation= 45)
plt.yticks(fontsize= 14)
plt.title("Distribuição da Probabilidade", fontsize= 18)
plt.grid()

plt.subplot(1,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins=30, density= True)
plt.colorbar()
plt.xlabel("Idade", fontsize= 14)
plt.ylabel("Renda Mensal", fontsize= 14)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.title("Distribuição de Renda por Idade", fontsize= 18);
```

```python
#Construindo uma figura com 4 gráficos.
plt.figure(figsize= (15, 5))

plt.subplot(2,2,1)
plt.hist(df.MonthlyIncome, bins= 30, color="y", density= True, orientation= "horizontal", rwidth= .8)
plt.xlabel("Renda Mensal", fontsize= 14)
plt.ylabel("Probabilidade/Renda", fontsize= 14)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.title("Distribuição da Probabilidade", fontsize= 18)
plt.grid()

plt.subplot(2,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins= 30, density= True)
plt.colorbar()
plt.xlabel("Idade", fontsize= 14)
plt.ylabel("Renda Mensal", fontsize= 14)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.title("Distribuição da Renda por Idade", fontsize= 18)

plt.subplot(2,2,4)
plt.hist(df.age, bins= 30, color= "g", density= True, orientation= "vertical", rwidth= .8)
plt.xlabel("Idade",fontsize= 14)
plt.ylabel("Probabilidade/Idade", fontsize= 14)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.title("Distribuição da Idade", fontsize= 18)
plt.grid();
 
```

```python
# Como ficou um gráfico em cima do outro podemos aumentar o tamanho da nossa figura.
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
plt.hist(df.MonthlyIncome, bins= 30, color="y", density= True, orientation= "horizontal", rwidth= .8)
plt.xlabel("Renda Mensal", fontsize= 14)
plt.ylabel("Probabilidade/Renda", fontsize= 14)
plt.xticks(fontsize= 14, rotation= 45)
plt.yticks(fontsize= 14)
plt.title("Distribuição da Probabilidade", fontsize= 18)
plt.grid()

plt.subplot(2,2,2)
plt.hist2d(df.age, df.MonthlyIncome, bins= 30, density= True)
plt.colorbar()
plt.xlabel("Idade", fontsize= 14)
plt.ylabel("Renda Mensal", fontsize= 14)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.title("Distribuição de Renda por Idade", fontsize= 18)

plt.subplot(2,2,4)
plt.hist(df.age, bins= 30, color="g", density= True, orientation= "vertical", rwidth= .8)
plt.xlabel("Idade",fontsize= 14)
plt.ylabel("Probabilidade/Idade", fontsize= 14)
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
plt.title("Distribuição da Probabilidade", fontsize= 18)
plt.grid();
```
